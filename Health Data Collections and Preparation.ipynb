```python
cases_malaysia['cluster_import'] = cases_malaysia['cluster_import'].fillna(cases_malaysia['cluster_import'].mode()[0])
cases_malaysia['cluster_religious'] = cases_malaysia['cluster_religious'].fillna(cases_malaysia['cluster_religious'].mode()[0])
cases_malaysia['cluster_community'] = cases_malaysia['cluster_community'].fillna(cases_malaysia['cluster_community'].mode()[0])
cases_malaysia['cluster_highRisk'] = cases_malaysia['cluster_highRisk'].fillna(cases_malaysia['cluster_highRisk'].mode()[0])
cases_malaysia['cluster_education'] = cases_malaysia['cluster_education'].fillna(cases_malaysia['cluster_education'].mode()[0])
cases_malaysia['cluster_detentionCentre'] = cases_malaysia['cluster_detentionCentre'].fillna(cases_malaysia['cluster_detentionCentre'].mode()[0])
cases_malaysia['cluster_workplace'] = cases_malaysia['cluster_workplace'].fillna(cases_malaysia['cluster_workplace'].mode()[0])
```

```python
cases_malaysia = cases_malaysia.dropna()
cases_malaysia.describe()
```
